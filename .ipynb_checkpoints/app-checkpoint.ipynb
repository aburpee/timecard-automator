```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
DRIVER_PATH = '/usr/local/bin/chromedriver'

first_employer = 'automedia'
dads_birthday = '04/18'
sibling_middle = 'gregory'
street_name = 'ashington'
prom = 'paulina'




driver = webdriver.Chrome()

driver.get('https://www.paycomonline.net/v4/ee/web.php/app/login')

driver.find_element(By.ID, 'txtlogin').send_keys('0LC63A046')
time.sleep(1)
driver.find_element(By.ID, 'txtpass').send_keys('')
time.sleep(3)
driver.find_element(By.ID, 'userpinid').send_keys('')
time.sleep(2)
driver.find_element(By.ID, 'btnSubmit').click()

time.sleep(5)


first_sec_question = driver.find_element(By.NAME, 'first_security_question').get_attribute('aria-label')
second_sec_question = driver.find_element(By.NAME, 'second_security_question').get_attribute('aria-label')

current_url = driver.current_url

if 'security' in current_url:

    if 'employer' in first_sec_question.lower():
        first_answer = first_employer
    elif 'birthday' in first_sec_question.lower():
        first_answer = dads_birthday
    elif 'sibling' in first_sec_question.lower():
        first_answer = sibling_middle
    elif 'street' in first_sec_question.lower():
        first_answer = street_name
    elif 'prom' in first_sec_question.lower():
        first_answer = prom
        
    time.sleep(5)

    driver.find_element(By.NAME, 'first_security_question').send_keys(first_answer)

    if 'employer' in second_sec_question.lower():
        second_answer = first_employer
    elif 'birthday' in second_sec_question.lower():
        second_answer = dads_birthday
    elif 'sibling' in second_sec_question.lower():
        second_answer = sibling_middle
    elif 'street' in second_sec_question.lower():
        second_answer = street_name
    elif 'prom' in second_sec_question.lower():
        second_answer = prom
    
    time.sleep(2)

    driver.find_element(By.NAME, 'second_security_question').send_keys(second_answer)
    
    time.sleep(1)

    driver.find_element(By.NAME, 'continue').click()
    
```
